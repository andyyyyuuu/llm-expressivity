```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path

def plot_results(title: str, csv_paths: dict[str, str], num_rows: int, num_cols: int, xmax: float = 12, ymax: float = 7): 
    fig, axs = plt.subplots(num_rows, num_cols, figsize=(num_cols * 4, num_rows * 4))
    
    fig.suptitle(title)
    axs = axs.flatten()

    path_prefix = Path('../data/')

    for i, (path, title) in enumerate(list(csv_paths.items())):
        df = pd.read_csv(path_prefix / path)
        axs[i].set(xlim=(0, 12), ylim=(0, 7), xlabel="Target $H$", ylabel="Best $D_{KL}$", title=title)
        axs[i].scatter(df['target_entropy'], df['best_loss'], s=1)

    for j in range(i + 1, len(axs)): axs[j].set_axis_off() # hide remaining, if any
    fig.tight_layout(rect=(0, 0, 1, 1))
    plt.show()
    
```

```python
plot_results(
    title="$\\it{Distribution~Prompting}$ Intervention Tuned at Different Locations\n(units in nats)",
    csv_paths={
        'embed.csv': 'After Embedding',
        'l7.csv': 'After Layer 7',
        'l15.csv': 'After Layer 15',
        'postnorm.csv': 'After Final RMSNorm'
    },
    num_rows=2,
    num_cols=2
)

# results/figures/vanilla.png
```

```python
plot_results(
    title="$\\it{Distribution~Prompting}$ on Tuned Reachable Distributions\n(units in nats)",
    csv_paths={
        'embed_reachable.csv': 'After Embedding',
        'postnorm_reachable.csv': 'After Final RMSNorm'
    },
    num_rows=1,
    num_cols=2
)

# results/figures/reachable.png
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from mpl_toolkits.axes_grid1.inset_locator import mark_inset

df = pd.read_csv('../data/linsoftmax.csv')
fig, ax = plt.subplots(figsize=(6.4, 4.8))
ax.scatter(df['target_entropy'], df['best_loss'], s=1.5)
ax.set(xlim=(0, 12), ylim=(0, 70), xlabel="Target $H$", ylabel="Best $D_{KL}$",
       title="$\\it{Distribution~Prompting}$ on Randomly Initialized Linear + Softmax\n(units in nats)")

ax.axhspan(0, 7, color="orange", alpha=0.08)

axins = ax.inset_axes([0.6, 0.4, 0.45, 0.45])
axins.axhspan(0, 7, color="orange", alpha=0.08)
axins.scatter(df['target_entropy'], df['best_loss'], s=0.5)
axins.set(xlim=(0, 12), ylim=(0, 7), xlabel="Target $H$", ylabel="Best $D_{KL}$")
mark_inset(ax, axins, loc1=2, loc2=4, fc="none", ec="lightgray")
plt.show()

# results/figures/linsoftmax.png
```
